## Question 1

```python
import pandas as pd
import numpy as np
```

### load data

```python
matrix_1 = pd.read_table('/content/drive/MyDrive/528_HW/matrix1.txt')
del matrix_1['Unnamed: 13']
matrix_1 = matrix_1.set_index('miRNA')
matrix_1.head()
```

### calculate correlation

```python
correlation_matrix = matrix_1.corr()
correlation_matrix
```

# Question 2

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Heatmap for matrix 1

```python
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Pearson Correlation Matrix 1 Heatmap', size=15)
```

### calculate correlation for matrix 2
